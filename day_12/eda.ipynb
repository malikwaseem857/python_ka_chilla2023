```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Import dataset

```python
#  data import
df = sns.load_dataset('titanic')
df
```

# 2. Bigger picture of data

```python
# how to check the data types
df.dtypes
```

```python
# it gives us the missing values of each column
df.info()
```

```python
# to calculate the null values or missing values in the dataset
df.isnull().sum()
```

```python
df2 = df
```

```python
df2
```

```python
# how to calculate the percentage of missing values
df2.isnull().sum()*100 / len(df2)
```

### Its a general rule that if a column has about 70-75% of missing values then we can drop that column
### for 2-4% of missing values we can impute them with mean, median or mode depending on the distribution of data rather than dropping them
### or we remove them if they are not that much important for our analysis

### for age column we can impute with mean or median
### for embarked column we can impute with mode
### for deck column we can drop that column as it has more than 70% missing values

```python
# how to find the unique values in a column
df2['age'].unique()
```

```python

# to find the maximum age
df2['age'].max()
```

```python
# to find the minimum age 
df2['age'].min()
```

```python
# to find the mean of age
df2['age'].mean()
```

```python
# how to replace the Nan values with mean
df2['age'].fillna(df2['age'].mean(), inplace=True)
```

```python
df2
```

```python
# now check the null values again
df2.isnull().sum()
```

```python
df3 = df2
```

```python
df3.isnull().sum()
```

```python
# to drop the deck column
df3.drop(columns=['deck'])
```

```python
# lets update the df3
df3 = df3.drop(columns=['deck'])
```

```python
# lets check the deck column is dropped or not
df3.isnull().sum()
```

```python
df4 = df3
```

```python
# lets replace the missing values in the embarked column with the mode
df4['embarked'].fillna(df4['embarked'].mode()[0], inplace=True)
```

```python
# lets check the null values again
df4.isnull().sum()
```

```python
# now lets replace the embark_town missing values with mode
df4['embark_town'].fillna(df4['embark_town'].mode()[0], inplace=True)
```

```python
# lets check the null values again
df4.isnull().sum()
```

```python
#lets check the mean of age in df
print("mean = ", df['age'].mean())
print("median = ", df['age'].median())
print("mode = ", df['age'].mode()[0])
```

```python
# how to check the  mode of embarked column
print("mode = ", df['embarked'].mode()[0])
```
