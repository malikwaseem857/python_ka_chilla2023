```python
# import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
kashti.info() # with () is a function without () is a attribute.
```

```python
# in 1st code just canvas has made 2nd code we scatter all the things w=white, 
g = sns.FacetGrid(kashti, hue="survived", col="pclass", margin_titles=True,
                  palette={1:"seagreen", 0:"grey"})
g=g.map(plt.scatter, "fare", "age", edgecolor= "w").add_legend()
```

```python
g = sns.FacetGrid(kashti, hue="survived", col="who", margin_titles=True,
                  palette={1:"seagreen", 0:"grey"})
g=g.map(plt.scatter, "fare", "age", edgecolor= "w").add_legend()
```

```python
# strip plot combined form of strip and bar plot
sns.stripplot(data= kashti, x="sex", y="fare", hue="class")
```

```python
#Heatmap   first we need co-relation co-relation value bw -1 to 1. 
corr = kashti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
```

```python
corr = kashti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("kashti ka heatmap")
```
