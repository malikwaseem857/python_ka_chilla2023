## KNN Grid Search CV

```python
# KNN Neighbors
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Load dataset
df = sns.load_dataset('titanic')

# Features and target
X = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y = df['survived']

# Encode and handle missing values
X = pd.get_dummies(X, columns=['sex'], drop_first=True)
X['age'].fillna(X['age'].mean(), inplace=True)

# Create the model
model = KNeighborsClassifier()

# Define parameter grid
param_grid = {'n_neighbors': [3, 5, 7, 9]}

# Grid search with cross-validation
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='f1')

# Fit the model
grid_search.fit(X, y)

# Print results
print("Best Parameters:", grid_search.best_params_)
print("Best F1 Score:", grid_search.best_score_)
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Load dataset
df = sns.load_dataset('titanic')

# Features and target
X = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y = df['survived']

# Encode and handle missing values
X = pd.get_dummies(X, columns=['sex'], drop_first=True)
X['age'].fillna(X['age'].mean(), inplace=True)

# Create the model
model = KNeighborsClassifier()

# Define parameter grid (fixed np.arange)
param_grid = {'n_neighbors': np.arange(1, 40, 2), 'weights': ['uniform', 'distance']}

# Grid search with cross-validation
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='recall')

# Fit the model
grid_search.fit(X, y)

# Print results
print("Best Parameters:", grid_search.best_params_)
print("Best Recall Score:", grid_search.best_score_)
```

## Decision Tree Classifier (DTC)

```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Load dataset
df = sns.load_dataset('titanic')

# Features and target
X = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y = df['survived']

# Encode and handle missing values
X = pd.get_dummies(X, columns=['sex'], drop_first=True)
X['age'].fillna(X['age'].mean(), inplace=True)

# Create the model
model = DecisionTreeClassifier()

# Define parameter grid (fixed np.arange)
param_grid = {'max_depth': [3,5,7,None], 'min_samples_split':[2,3,4]}

# Grid search with cross-validation
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='recall')

# Fit the model
grid_search.fit(X, y)

# Print results
print("Best Parameters:", grid_search.best_params_)
print("Best Recall Score:", grid_search.best_score_)
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Load dataset
df = sns.load_dataset('titanic')

# Features and target
X = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y = df['survived']

# Encode and handle missing values
X = pd.get_dummies(X, columns=['sex'], drop_first=True)
X['age'].fillna(X['age'].mean(), inplace=True)

# Create the model
model = RandomForestClassifier()

# Define parameter grid (fixed np.arange)
param_grid = {'n_estimators': [10, 50, 100],
              'max_depth': [3,5,7, None],
              'min_samples_split':[2,3,4],
              'max_features': ['auto, sqrt', 'log2']}

# Grid search with cross-validation
grid_search = GridSearchCV(model, param_grid, cv=5, scoring= 'precision')

# Fit the model
grid_search.fit(X, y)

# Print results
print("Best Parameters:", grid_search.best_params_)
print("Best Recall Score:", grid_search.best_score_)
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Load dataset
df = sns.load_dataset('titanic')

# Features and target
X = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y = df['survived']

# Encode and handle missing values
X = pd.get_dummies(X, columns=['sex'], drop_first=True)
X['age'].fillna(X['age'].mean(), inplace=True)

# Create the model
model = RandomForestClassifier(random_state=42)

# Correct parameter grid
param_grid = {
    'n_estimators': [10, 50, 100],
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 3, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Grid search with cross-validation
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='precision')

# Fit the model
grid_search.fit(X, y)

# Print results
print("Best Parameters:", grid_search.best_params_)
print("Best Precision Score:", grid_search.best_score_)
```
