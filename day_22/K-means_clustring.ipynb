## K-Means clustering

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
df.head()
```

```python
# Create a model

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


# Import the data in features
X = df[['age', 'fare']]

# scaled data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X.head()
```

```python
# scaling is from -3 to +3
X_scaled
```

```python
# Create a model

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


# Import the data in features
X = df[['age', 'fare']]

# Encode and handle missing values
X['age'].fillna(X['age'].mean(), inplace=True)

# scaled data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# perform clustering
model = KMeans(n_clusters=3)
model.fit(X_scaled)

# assign labels
df['cluster']= model.labels_

```

```python
# we add cluster in the end
df.head()
```

```python
# lets check the unique value
df['cluster'].unique()
```

```python
# lets change just the cluster
# Create a model
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


# Import the data in features
X = df[['age', 'fare']]

# Encode and handle missing values
X['age'].fillna(X['age'].mean(), inplace=True)

# scaled data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# perform clustering
model = KMeans(n_clusters=5)
model.fit(X_scaled)

# assign labels
df['cluster']= model.labels_
```

```python
# lets check the unique value again
df['cluster'].unique()
```

```python
# lets create plot

plt.scatter(df['age'], df['fare'])
plt.show()
```

```python
# lets create plot and separate with cluster

plt.scatter(df['age'], df['fare'], c=df['cluster'])
plt.show()
```

```python
# lets create plot and separate with cluster and cmap rainbow

plt.scatter(df['age'], df['fare'], c=df['cluster'], cmap='rainbow')
plt.show()
```

```python
plt.scatter(df['age'], df['fare'], c=df['cluster'], cmap='rainbow')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], color='black', marker='X', s=200)
plt.show()
```

```python
plt.scatter(df['age'], df['fare'], c=df['cluster'], cmap='rainbow')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], color='black', marker='X', s=200, label='Centroids')
plt.xlabel('age')
plt.ylabel('fare')
plt.legend()
plt.show()
```
