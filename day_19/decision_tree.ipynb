## Decision Tree Classification

#### Import libraries

```python
import pandas as pd
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
df = sns.load_dataset('titanic')
```

```python
# lets take 6 columns for x, and 1 column for y
x = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y = df['survived']
```

```python
# let get dummies done with pandas library and get dummy function as well.
x = pd.get_dummies(x, columns=['sex'], drop_first=True)
```

```python
x.isnull().sum()
```

```python
# lets fill the null values with mean of age column
x['age'].fillna(x['age'].mean(), inplace=True)
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# lets train the decision tree model
# in day_18, we have done linear regression model
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
# predict on test data
y_pred = model.predict(x_test)
```

```python
# lets check the accuracy of the model
print("Accuracy", accuracy_score(y_test, y_pred))
```

```python
# lets print decision tree
tree.plot_tree(model.fit(x, y), filled=True)
plt.show()
plt.savefig('decision_tree.png')
```

### K Neartest Neighbour (KNN)

```python
# lets import the Machine Learning Model and Libraries

import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score


from sklearn.linear_model import LogisticRegression
from sklearn.neighbors  import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# define the model to be evaluate
models = [LogisticRegression(), KNeighborsClassifier(), RandomForestClassifier(), SVC()]
names = ["Logistic Regression","KNN", "Random Forest", "SVC"]

# Perform k-fold cross_validation for each model
k= 10
for name, model in zip(names, models):
    cv_scores = cross_val_score(model, x_train, y_train, cv=k)
    print(f"{name} CV accuracy: { np.mean(cv_scores):.3f} +/- {np.std(cv_scores):.3f}")


```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10],        # capital 'C' — parameter name in SVC
    'kernel': ['linear', 'rbf'],
    'gamma': [0.1, 1, 10]     # correct spelling: 'gamma'
}

# Create the GridSearchCV object
grid_search = GridSearchCV(SVC(), param_grid, cv=5, verbose=2, n_jobs=-1)

# Fit the model
grid_search.fit(x_train, y_train)

# Print the best parameters and best model
print("Best Parameters:{}" .format(grid_search.best_params_))
print("Best cross-validation score: {:2f}".format(grid_search.best_score_))

# this is called hypermeter tunning 
```
