## Decision Tree Classification

#### Import libraries

```python
import pandas as pd
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
df = sns.load_dataset('titanic')
```

```python
# lets take 6 columns for x, and 1 column for y
x = df[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y = df['survived']
```

```python
# let get dummies done with pandas library and get dummy function as well.
x = pd.get_dummies(x, columns=['sex'], drop_first=True)
```

```python
x.isnull().sum()
```

```python
# lets fill the null values with mean of age column
x['age'].fillna(x['age'].mean(), inplace=True)
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# lets train the decision tree model
# in day_18, we have done linear regression model
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
# predict on test data
y_pred = model.predict(x_test)
```

```python
# lets check the accuracy of the model
print("Accuracy", accuracy_score(y_test, y_pred))
```

```python
# lets print decision tree
tree.plot_tree(model.fit(x, y), filled=True)
plt.show()
plt.savefig('decision_tree.png')
```

```python
# precision 
# if your all predictions are correct, then precision is 1.0
# if your all predictions are wrong, then precision is 0.0

from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_test, y_pred)

```

### Precision : Precision measure the proportion of true positive prediction among all positive instances.

```python
# precision 
from sklearn.metrics import precision_score, recall_score, f1_score

precision_score(y_test, y_pred)
```

### Recall: Recall measure the proportion of true positive prediction among all actuall instances.

```python
# lets recall score
recall_score(y_test, y_pred)
```

```python
# lets recall f1 score
# f1 score is the harmonic mean of precision and recall
```

### K Neartest Neighbour (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
# if you will change the neighbour values will be changes
# hyper
model = KNeighborsClassifier(n_neighbors=3)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
cn = confusion_matrix(y_test, y_pred)
cn
```

```python
# lets check the precision score
precision_score(y_test, y_pred)
```

```python
# lets check the recall score
recall_score(y_test, y_pred)
```

```python
sns.heatmap(cn, annot=True)
plt.ylabel('True label')
plt.xlabel('predicted label')
```
