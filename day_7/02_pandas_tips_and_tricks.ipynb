```python
df.rename(columns={'A col': 'col_a', 'B col': 'col_b'}, inplace=True)  # rename columns
df
```

```python
# rename columns
df.columns=['col_aa', 'col_ba']
df
```

```python
# rename columns
df.columns=['col_aa', 'col_bb']
df
```

```python
# to replace any character , string
df.columns=df.columns.str.replace('_','') # original first, require second will happen
df
```

```python
# how to do the reverse of the above one
df.columns=df.columns.str.replace(' ','_')
df
```
