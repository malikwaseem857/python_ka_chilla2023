# Exploratory Data Analysis:
Exploratory Data Analysis (EDA) is a method to `analyze` and `summarize` datasets. In python, you can use the pandas to perform EDA on a ddataset.


```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.tail()
```

```python
# shape and data type  is inside the info
df.info()
```

```python
# summarize the data (summar statistics) of numerical columns
df.describe()
```

```python
# find missing values or null values
df.isnull()
```

```python
# to get total missing values in each column
df.isnull().sum()
```

```python
# to get the percentage of missing values in each column
df.isnull().mean()/100/len(df) *100
```

```python
# to get the number of  unique values in a column
df.nunique()
```

### if a value is more than 10 numbers then it will be usually numerical value some of them are categorical

```python
df.columns
```

```python
# visualization of missing values in a heatmap
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=True, cmap='viridis')
```

```python
# to drop the deck column
df.drop('deck', axis=1, inplace=True)
```

```python
df.columns
```

```python
df.shape
```

```python
# visualize and explore the data
# for count plot no need to insert the details of y-axis because it already mentioned in the count plot
sns.countplot(data= df, x='survived')
```

```python
# for barplot
sns.barplot(data=df, x='survived', y='age', hue= 'who')
```

```python
# for barplot
sns.barplot(data=df, x='survived', y='fare', hue= 'who')
```

```python
sns.countplot(x='pclass', data=df)
```

```python
# data distribution plot
sns.histplot(data=df, x='age')
```

```python
sns.histplot(data=df, x='age', bins=20)
```

```python
sns.histplot(data=df, x='age', bins=5)
```

```python
sns.histplot(data=df, x='age', bins=10)
```

```python
df['age'].hist(bins=30, color='darkred', alpha=0.7)
```

```python
# this is useless because seaborn boxplot need categorical data on x-axis
sns.boxplot(data=df, x='age', y='pclass')
```

```python
sns.boxplot(data=df, x='survived', y='age')
```

```python
sns.barplot(data=df, x='survived', y='fare', hue= 'who')
```

```python
sns.boxplot(data=df, x='survived', y='age', hue='who')
```

```python
# scatter plot
sns.scatterplot(data=df, x='age', y='fare')
plt.ylim(0, 300)
plt.show()
```

```python
# scatter plot
sns.scatterplot(data=df, x='age', y='fare', hue='pclass')
plt.ylim=(0,300)
```

```python
# pair plot
sns.pairplot(data=df, hue='survived')
```

```python
# stacked bar chart/plot
df.groupby(['pclass', 'sex']).size().unstack().plot(kind='bar', stacked=True)
```

```python
# stacked bar chart/plot
df.groupby(['pclass', 'who']).size().unstack().plot(kind='bar', stacked=True)
```

```python
# stacked bar chart/plot
df.groupby(['pclass', 'who'])['fare'].size().unstack().plot(kind='bar', stacked=True)
```

```python
# swarm plot
sns.swarmplot(data=df, x='survived', y ='fare')
```

```python
# swarm plot
sns.swarmplot(data=df, x='survived', y ='age')
```

```python
# swarm plot
sns.swarmplot(data=df, x='survived', y ='age', hue='who')
```

```python
# point plot
sns.pointplot(data=df, x='pclass', y='fare', hue='embarked')
```

```python
# LM plot for relation between two variables
sns.lmplot(data=df, x='age', y='fare')
```

```python
sns.lmplot(data=df, x='age', y='fare', hue='pclass')
```

```python
# how to add a column and make a age group
sns.histplot(data=df, x='age')
```

```python
# grouping
df.columns
```

```python
# lets add age group
df['age_groups'] = pd.cut(df['age'], bins=[0,10,18,30,50,80], labels=['child','teenager','adult','middle_aged','senior'])
```

```python
df.head()
```

```python
df.groupby('age_groups')['survived'].mean().plot(kind='bar')
```

```python
# pair plot
sns.pairplot(data=df, hue='survived', vars=['age','fare','pclass'])
```
